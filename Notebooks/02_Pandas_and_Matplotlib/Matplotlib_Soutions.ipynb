Given a string, return a new string where the first and last characters have been exchanged.

```python
#write code here
def swap_chars(old_string):
    a = old_string[0]
    b = old_string[len(old_string)-1:]
    new_string = b + old_string[1:-1] + a
    return new_string
    if len(old_string) == 1:
      return old_string 
```
